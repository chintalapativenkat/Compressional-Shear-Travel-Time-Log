## Creating a Pipeline for Automating Feature Transformation and Model Training

```python
def reset_index(dataframe):
    dataframe=dataframe.reset_index(inplace=False,drop=True)
    return dataframe
```

```python
get_reset_index= FunctionTransformer(reset_index,validate=False)
```

```python
def replace_999_null(dataframe):
    dataframe= dataframe.replace(['-999', -999], np.nan, inplace=False)
    dataframe=dataframe.dropna(axis=0, inplace=False)
    return dataframe
```

```python
replace_999_values = FunctionTransformer(replace_999_null,validate=False)
```

```python
def outlier_using_isolation_forest(dataframe):
    model_iso=IsolationForest(contamination=outliers_fraction)
    model_iso.fit(dataframe)
    dataframe['bool_outlier'] = pd.Series(model_iso.predict(dataframe))
    dataframe['bool_outlier'] = dataframe['bool_outlier'].map( {1: 0, -1: 1})
    dataframe=dataframe[dataframe['bool_outlier']==0]
    dataframe=dataframe.drop(['bool_outlier'],axis=1)
    return dataframe
```

```python
outlier_removal= FunctionTransformer(outlier_using_isolation_forest,validate=False)
```

```python
feature_transform_pipe_model=Pipeline(memory=None, 
            steps=[('replace_null_values',replace_999_values),
                   ('reset_index_of_dataframe_1',get_reset_index),
                   ('outlier_detection',outlier_removal),
                   ('reset_index_of_datframe_2',get_reset_index)
                  ],verbose=False)
```

```python
df_transformed= feature_transform_pipe_model.fit_transform(df)
```

```python
df_transformed
```
